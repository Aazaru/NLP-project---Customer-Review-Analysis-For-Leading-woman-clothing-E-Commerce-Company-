```python
import string
import re
import contractions
import nltk
from nltk.stem import PorterStemmer,WordNetLemmatizer
ps=PorterStemmer()
lm=WordNetLemmatizer()
```

```python
stopward=nltk.corpus.stopwords.words("english")
stopward=list(set(stopward+["the","me","how","what","wa",'xsi', 'xslbs', 'yo',"x","ha"]))
```

```python
def lower(x):
    return x.lower()

def strip(x):
    return x.strip()

def contraction(x):
    return contractions.fix(x)

def special_char_digit(x):
    x = re.sub(r" +", " ", x)
    x = re.sub("[^a-z\s]", " ", x)
    x = re.sub("\{2,}", " ", x)
    x = re.sub(r"\d+", " ", x)
    return x

def punctuation(x):
    return ''.join([c for c in x if c not in string.punctuation])

def stemming(x):
    return [ps.stem(word) for word in re.findall(r"\b\w+\b", x)]

def lemmatizing(x):
    return [lm.lemmatize(word) for word in re.findall(r"\b\w+\b", x)]

def stopword_(x):
    return ' '.join([word for word in re.findall(r"\b\w+\b", x) if word not in stopward])

def remove_repeated_words(x):
    words = x.split()  
    unique_words = set(words) 
    cleaned_text = ' '.join(unique_words)  
    return cleaned_text
```

```python
def clean_data(x, baseform):
    x = lower(x)
    x = strip(x)
    x = special_char_digit(x)
    x = punctuation(x)
    if baseform == "stem":
        x = ' '.join(stemming(x))
    else:
        x = ' '.join(lemmatizing(x))
    x = stopword_(x)
    x= remove_repeated_words(x)
    return x
```

```python
data["clean_text"]=data['Review Text'].apply(lambda y: clean_data(y,"lemm"))
```

```python
from gensim import corpora
from gensim.models import LdaModel
```

```python
doc=data["clean_text"].tolist()
```

```python
document=[i.split() for i in doc]
```

```python
dictionary=corpora.Dictionary(document)
corpus=[dictionary.doc2bow(i) for i in document]
lda=LdaModel(corpus,num_topics=5,id2word=dictionary,passes=30)
topic=lda.show_topics(num_topics=5,num_words=30,formatted=False)
```

```python
for topic_id,topic_value in topic:
    print(f"topic:{topic_id+1}")
    print(", ".join(i[0] for i in topic_value))
```
